## **1. üî¨ Referencias Fundacionales del Modelo**

### **Modelos Matem√°ticos Base:**

UXsim est√° basado en modelos te√≥ricos validados y ampliamente utilizados en investigaci√≥n acad√©mica de transporte:

### **üìñ Newell's Car-Following Model (2002)**

**Referencia Principal:**
- Newell, G. F. (2002). [A simplified car-following theory: a lower order model](https://doi.org/10.1016/S0191-2615(00)00044-8). *Transportation Research Part B: Methodological*, 36(3), 195-205.

**Importancia:** Modelo fundamental de seguimiento vehicular simplificado pero riguroso matem√°ticamente.

**Ecuaci√≥n clave:** `v = min(u, (x_leader - x - Œ¥)/œÑ)`
- `v`: velocidad del veh√≠culo
- `u`: velocidad libre
- `Œ¥`: espaciado m√≠nimo
- `œÑ`: tiempo de reacci√≥n

### **üìñ X-Model Extension (2013)**

**Referencia:**
- Laval, J. A., & Leclercq, L. (2013). [The Hamilton‚ÄìJacobi partial differential equation and the three representations of traffic flow](https://doi.org/10.1016/j.trb.2013.02.008). *Transportation Research Part B: Methodological*, 52, 17-30.

**Mejora:** Manejo m√°s preciso de ondas de congesti√≥n y transiciones de estado del tr√°fico.

### **üìñ Incremental Node Model (2001)**

**Referencia:**
- Peeta, S., & Ziliaskopoulos, A. K. (2001). [Foundations of dynamic traffic assignment: The past, the present and the future](https://doi.org/10.1016/j.trb.2011.04.001). *Networks and Spatial Economics*, 1(3-4), 233-265.

**Versi√≥n Mesosc√≥pica:**
- Himpe, W., & Verbeke, W. [Mesoscopic traffic simulation](https://ubiquitypress.com/site/chapters/e/10.5334/baw.50/). *Transportation Research Arena*.

**Funci√≥n:** Modelo de intersecciones que preserva principios de conservaci√≥n de flujo.

### **üìñ Dynamic User Equilibrium (2001)**

**Referencia:**
- Friesz, T. L., et al. (2001). [A variational inequality formulation of the dynamic network user equilibrium problem](https://doi.org/10.1016/S0191-2615(00)00005-9). *Transportation Research Part B: Methodological*, 35(3), 179-196.

**Aplicaci√≥n:** Asignaci√≥n din√°mica de tr√°fico con equilibrio de usuario.
